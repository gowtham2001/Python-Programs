```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("/content/Latest Covid-19 India Status.csv")
df.shape
df.head(20)
```

### **TOTAL CASES ACROSS STATES**

```python
plt.figure(figsize=(20,10))
sns.barplot(x='State/UTs',y='Total Cases',data=df)
plt.xticks(rotation=90)
```

# **ACTIVE CASES**

```python
plt.figure(figsize=(15,10))
sns.pointplot(data=df, x="State/UTs", y="Deaths")
plt.xticks(rotation=90)
plt.show()
```

# **DISCHARGED CASES**

```python
plt.figure(figsize=(15,10))
sns.lineplot(x='State/UTs',y='Discharged',data=df)
plt.xticks(rotation=90)
plt.show()
```

# **DEATHS ACROSS STATES**

```python
plt.figure(figsize=(15,10))
plt.scatter(x='State/UTs',y='Deaths',data=df)
plt.xticks(rotation=90)
plt.show()
```
